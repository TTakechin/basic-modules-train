### モジュール　Matplotlib.pyplotの使い方

---

Matplotlibは２D、３Dグラフを描くためのモジュールである。そして、Matplotlib.pyplotはMaplotlibのメソッドの一つであるが、これだけでもかなり十分な機能はあり、通常の科学技術系のグラフを描くのであればこれで十分である。

【Matplotlib.pyplotでできること】（代表例）

1. 2Dグラフの作成
2. 3Dグラフの作成
3. グラフのレイアウト

Link to: [Matplotlib official page](https://matplotlib.org)

```python
### モジュールのインポート
import matplotlib.pyplot as plt

%matplotlib inline 
#Jupyterlab/Jupyternotebook特有。セルの間にうまく画像をはめ込んでくれる
```

まず覚えるのは４種類のグラフ：
1. 散布図（scatter）
2. 線図（plot）
3. 棒グラフ(bar)
4. ヒストグラム（hist）

ヒストグラムは馴染みが無いかもしれませんが、統計分野（データサイエンスを実施する上で必須の分野）では、基本中の基本くらいの勢いで出てきます。ヒストグラムが表すものは、存在しているデータがどの範囲にどれだけあるか、ということです。詳しくはPandasのところで説明しますので、このここの回ではグラフの描き方のみ覚えておけばOKです。

---
#### 散布図の描き方

同じ個数のXとYを使って描くことができる


```python
X = [1,2,3,4,5]
Y = [6,8,8,7,10]
```

```python
plt.scatter(X,Y)
```

---
#### 線図の描き方

基本的には散布図と描き方は同じ

```python
plt.plot(X,Y)
```

---
#### 棒グラフの描き方

こちらも、基本的には散布図と同じ

```python
plt.bar(X,Y)
```

棒グラフ特有かもしれませんが、ラベルをつけることもできます。

```python
label = ["red","green","blue","orange","purple"]
```

```python
plt.bar(X,Y)
plt.xticks(X,label,rotation=0) #同じセル内に書きます。
plt.xlabel("fruit")
plt.ylabel("counts")
```

```python
plt.hist(Y)
```

---

#### グラフの装飾

X、Yラベル、凡例も追加することができます。

```python
plt.title("Test figure") #タイトルをつける
plt.scatter(X,Y,label="Fruits") #散布図プロットを作成
plt.xlabel("X value") #X軸のラベルをつける
plt.ylabel("Y value") #Y軸のラベルをつける
plt.legend() #凡例をつける
```

```python
plt.figure(figsize=(10,8)) #グラフのサイズを指定する
plt.title("Test figure")
plt.scatter(X,Y,label="Fruits",s=300,alpha=0.4) #s:サイズ、alpha:透明度
plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend()
```

---

数学のグラフの描き方：

1. xの値を準備する
2. 描きたい関数を作る（注：ここでの関数は数学の関数、defでは無い）
3. 計算結果をもとにグラフを描く

例) y = x**2 - 6　のグラフを-3<=x<=6の範囲で描きなさい

```python
import numpy as np
```

```python
xsim = np.linspace(-3,6,100)
```

```python
ysim = xsim**2 - 6
```

```python
plt.figure()
plt.plot(xsim,ysim,color="red")
plt.xlabel("X")
plt.ylabel("Y")
```
