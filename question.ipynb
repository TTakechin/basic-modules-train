数字生成

```python
#1から100まで刻んで表示する
np.linspace(1, 10, 100)
```

```python
# 7^1から7^3まで100刻んでを表示する
np.logspace(1, 3, 100, base=7)
```

```python
a = np.linspace(3, 40, 200)
a
```

```python
a = a * 5
a
```

```python
#平均値
np.average(a)
```

```python
#分散値
np.var(a)
```

```python
#標準偏差
np.std(a)
```
